```python
from __future__ import division
import pandas as pd
import numpy as np
import chemfun as cf
from numpy.linalg import eigvalsh,eigh
from rdkit import Chem
from rdkit.Chem.rdmolops import GetAdjacencyMatrix
from rdkit.Chem import AllChem
from rdkit.Chem.Crippen import MolLogP,MolMR
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel as C
from sklearn.gaussian_process.kernels import Matern, WhiteKernel
# #import scipy as sp
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import networkx as nx
from numpy import linalg as LA
% matplotlib inline
```

```python
# Importing data from txt file into numpy arrays
data=pd.read_table("delaney.txt",header=None)
smiles=data[0].values
logS=data[1].values

# Setting indices to know how data is split 
indices=np.arange(len(smiles))

#Splitting data into train/test sets
Train,Test,TrainTargets,TestTargets,TrainInd,TestInd=train_test_split(smiles,logS,indices,test_size=0.2)#,random_state=20)
```

```python
## Comment these in to load saved train/test sets with their labels
Train=np.load('Train.npy')
Test=np.load('Test.npy')
TrainTargets=np.load('TrainTargets.npy')
TestTargets=np.load('TestTargets.npy')
```

### 1. Largest eigenvalue of adjacency matrix

```python
# Creating descriptors for train set 
ATR=[]
for mole in Train:
    A=cf.AdjMat(mole)
    ATR.append(cf.GetPrincipalEigenvalue(A))
    
# Creating descriptors for test set 
ATE=[]
for mole in Test:
    A=cf.AdjMat(mole)
    ATE.append(cf.GetPrincipalEigenvalue(A))
```

```python
# Turning lists into suitable numpy arrays
ATR=np.array(ATR).reshape(-1,1)
ATE=np.array(ATE).reshape(-1,1)
```

```python
# Fitting Gaussian process regression model
kernel=C()*RBF()#constant_value=1.0,constant_value_bounds=(1e-5,1e6))*RBF(length_scale=1.75,length_scale_bounds=(1e-5,1e6))
gp=GaussianProcessRegressor(kernel=kernel,normalize_y=False,n_restarts_optimizer=10)
gp.fit(ATR,TrainTargets)
TrainPred=gp.predict(ATR).reshape(-1,1)
TestPred=gp.predict(ATE).reshape(-1,1)
print('Kernel parameters are:',gp.kernel_)
```

```python
sse=cf.SSE(TestPred,TestTargets)
mse=sse*(1/len(Test))

print("Mean Squared Error (MSE) before deleting outliers = ", mse)
print("Sum Squared Error (SSE) before deleting outliers = ", sse)
```

```python
# Getting rid of outliers
lt=np.ravel(np.where(TestPred < -10))
gt=np.ravel(np.where(TestPred > 2))
delete=np.concatenate((lt,gt),0)
TestPred1=np.delete(TestPred,delete)
TestTargets1=np.delete(TestTargets,delete)

# Recalculating MSE,SSE
sse=cf.SSE(TestPred1,TestTargets1)
mse=sse*(1/len(TestPred1))

print("MSE after deleting outliers = ", mse)
print("SSE after deleting outliers = ", sse)
print('Number of Outliers with logS less than -10 or greater than 2:',len(delete))
```

```python
# Plot without outliers
fig=plt.figure(num=None,figsize=(8,6),dpi=500,facecolor='w',edgecolor='k')
plt.xlabel('Experimental',fontsize=15)
plt.ylabel('Prediction',fontsize=15)
plt.title('Model with Adjacency Matrix',loc='left',fontsize=18)
plt.scatter(TrainTargets,TrainPred,color='gold',alpha=1,label='Train',marker='D')
plt.scatter(TestTargets1,TestPred1,color='b',alpha=0.8,label='Test',facecolors='none')
plt.legend(fontsize=14,loc=4);
#plt.savefig('1.png');
```

### 2. Largest eigenvalue of bond-order matrix

```python
# Creating descriptors for train set 
# A few molecules do not have a bond order matrix that can be created
BTR=[]
delete_train=[]
i=0
for mole in Train:
    E=cf.GetEdges(mole)
    if len(E) == 1:
        delete_train.append(i)
    elif len(E) == 0:
        delete_train.append(i)
    else:
        B=cf.BondMat(mole)
        BTR.append(cf.GetPrincipalEigenvalue(B))
    i+=1

# Creating descriptors for test set 
BTE=[]
delete_test=[]
i=0
for mole in Test:
    E=cf.GetEdges(mole)
    if len(E) == 1:
        delete_test.append(i)
    elif len(E) == 0:
        delete_test.append(i)
    else:
        B=cf.BondMat(mole)
        BTE.append(cf.GetPrincipalEigenvalue(B))
    i+=1
```

```python
# Turning lists into suitable numpy arrays
BTR=np.array(BTR).reshape(-1,1)
BTE=np.array(BTE).reshape(-1,1)
```

```python
# Deleting deleted molecules from target values
TrainTargets1=np.delete(TrainTargets,delete_train)
TestTargets1=np.delete(TestTargets,delete_test)
```

```python
# Fitting Gaussian process regression model (same kernel)
gp=GaussianProcessRegressor(kernel=kernel,normalize_y=False,n_restarts_optimizer=0)
gp.fit(BTR,TrainTargets1)
TrainPred=gp.predict(BTR).reshape(-1,1)
TestPred=gp.predict(BTE).reshape(-1,1)
print('Kernel parameters are:',gp.kernel_)

# Getting rid of outliers
lt=np.ravel(np.where(TestPred < -10))
gt=np.ravel(np.where(TestPred > 2))
delete=np.concatenate((lt,gt),0)
TestPred1=np.delete(TestPred,delete)
TestTargets1=np.delete(TestTargets1,delete)

# Recalculating MSE,SSE
sse=cf.SSE(TestPred1,TestTargets1)
mse=sse*(1/len(TestPred1))

print("MSE after deleting outliers = ", mse)
print("SSE after deleting outliers = ", sse)
print('Number of Outliers with logS less than -10 or greater than 2:',len(delete))
```

```python
# Plot without outliers
fig=plt.figure(num=None,figsize=(8,6),dpi=500,facecolor='w',edgecolor='k')
plt.xlabel('Experimental',fontsize=15)
plt.ylabel('Prediction',fontsize=15)
plt.title('Model with Bond-order Matrix',loc='left',fontsize=18)
plt.scatter(TrainTargets1,TrainPred,color='gold',alpha=1,label='Train',marker='D')
plt.scatter(TestTargets1,TestPred1,color='b',alpha=0.8,label='Test',facecolors='none')
plt.legend(fontsize=14,loc=4);
#plt.savefig('2.png');
```

### 3. Largest eigenvalue of polarizability-weighted (colored) graph 

```python
# Loading polarizability data 
polarfile='Polars.txt'
polar=np.loadtxt(polarfile).astype(float)
AtomNumb=polar[:,0].astype(int)
Polar=polar[:,1]

# creating a dictionary to match polarizability to atomic number
df1=pd.DataFrame(data=AtomNumb,columns=["Atomic Number"])
df2=pd.DataFrame(data=Polar,columns=["Polarizability"])
PolarDF=pd.concat([df1,df2],axis=1,sort=False)
PolarDF=PolarDF.set_index('Atomic Number')
PolarDict=dict(zip(PolarDF.index,PolarDF.values))
```

```python
# Creating polar weighted  
PWTR=[]
for mole in Train:
    MFS=Chem.MolFromSmiles(mole)
    dim=MFS.GetNumAtoms()
    B=cf.GetColorMat(mole,dim,PolarDict,'Polar')
    PWTR.append(cf.GetPrincipalEigenvalue(B))
    
# Creating descriptors for test set 
PWTE=[]
for mole in Test:
    MFS=Chem.MolFromSmiles(mole)
    dim=MFS.GetNumAtoms()
    B=cf.GetColorMat(mole,dim,PolarDict,'Polar')
    PWTE.append(cf.GetPrincipalEigenvalue(B))
```

```python
PWTR=np.array(PWTR).reshape(-1,1)
PWTE=np.array(PWTE).reshape(-1,1)
```

```python
# Fitting Gaussian process regression model
gp=GaussianProcessRegressor(kernel=kernel,normalize_y=False,n_restarts_optimizer=0)
gp.fit(PWTR,TrainTargets)
TrainPred=gp.predict(PWTR).reshape(-1,1)
TestPred=gp.predict(PWTE).reshape(-1,1)
print('Kernel parameters are:',gp.kernel_)

# Getting rid of outliers
lt=np.ravel(np.where(TestPred < -10))
gt=np.ravel(np.where(TestPred > 2))
delete=np.concatenate((lt,gt),0)
TestPred1=np.delete(TestPred,delete)
TestTargets1=np.delete(TestTargets,delete)

# Recalculating MSE,SSE
sse=cf.SSE(TestPred1,TestTargets1)
mse=sse*(1/len(TestPred1))

print("MSE after deleting outliers = ", mse)
print("SSE after deleting outliers = ", sse)
print('Number of Outliers with logS less than -10 or greater than 2:',len(delete));
```

```python
# Plot without outliers
fig=plt.figure(num=None,figsize=(8,6),dpi=500,facecolor='w',edgecolor='k')
plt.xlabel('Experimental',fontsize=15)
plt.ylabel('Prediction',fontsize=15)
plt.title('Model with Polarizability Weighted Matrix',loc='left',fontsize=18)
plt.scatter(TrainTargets,TrainPred,color='gold',alpha=1,label='Train',marker='D')
plt.scatter(TestTargets1,TestPred1,color='b',alpha=0.8,label='Test',facecolors='none')
plt.legend(fontsize=14,loc=4);
#plt.savefig('3.png');
```
